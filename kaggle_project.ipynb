```python
def convert_rank_to_midpoint(rank):
    if isinstance(rank, str) and '-' in rank:
        start, end = map(int, rank.split('-'))
        return (start + end) / 2
    try:
        return float(rank)
    except ValueError:
        return np.nan

y = df['world_rank'].apply(convert_rank_to_midpoint)


y = pd.Series(SimpleImputer(strategy='mean').fit_transform(y.values.reshape(-1, 1)).flatten())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Model R^2: {model.score(X_test, y_test)}")
```
